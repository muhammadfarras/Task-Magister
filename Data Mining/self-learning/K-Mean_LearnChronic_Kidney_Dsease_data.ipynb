## Menentukan vector penentu dan target variable 

```python
y = df['Diagnosis']
df.drop(['Diagnosis'], inplace=True, axis=1)
X = df
```

## Feature Scalling

```python
from sklearn.preprocessing import MinMaxScaler
cols = X.columns
ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns= [cols])
```

```python
X.head()
```

```python
X.info()
```

# K-Means dengan dua cluster

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
```

## Telaah nilai model K-Means

```python
kmeans.cluster_centers_
```

```python
# Semakin sedikit tingkat inertia, maka semakin baik model tersebut
# Hasil kalkulasi jaraka antara centeroid dengan setiap titik pada data
kmeans.inertia_
```

## Check label

```python
labels = kmeans.labels_
labels.shape
```

```python
y.shape
```

```python
print('Jumlah data yg benar ', (sum(y==labels)/y.size*100))
```

## Optimasi jumlah banyak K yg dibutuhkan

Kita menginisiasi jumlah cluster menjadi 2 kelompok. untuk mengetahui berapa jumlah cluster K yang paling optimal menggunakan _Elbow Method_. Kita bisa mengukurnya dengan membuat plot X dan Y inertia dengan jumlah cluster

```python
cs = []
for nKBanyak in range (1,11):
    nKmeans = KMeans(n_clusters=nKBanyak, random_state=0)
    nKmeans.fit(X)
    cs.append(nKmeans.inertia_)

plt.plot (range(1,11),cs,'-')
plt.plot (range(1,11),cs,'o')
plt.xlabel('Banyak Cluster')
plt.ylabel('Intertia')
plt.title('Elbow Method')
plt.show()
```
