## Koefesien Regresei Berganda

```python

import statsmodels.api as sm
from scipy import statsimport pandas as pd
```

```python
## Preparing the data
pendapatan =  [10,10,8,5,4,3,2,2,1,1]
harga_minyak = [8,7,7,7,6,6,6,6,5,5]
permintaan = [3,4,5,6,6,7,8,9,10,10]

data = pd.DataFrame({"x1":harga_minyak,"x2":pendapatan, "y":permintaan})

data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = LinearRegression()

model.fit(data[['x1','x2']],data[['y']])

coeficient = model.coef_
intercept = model.intercept_
```

```python
coeficient,intercept
```

```python
print(f"mode regresi Y = {intercept[0]} + (harga_minyak * {coeficient[0][0]}) + (pendapatan * {coeficient[0][1]})")
```

```python
X2 = sm.add_constant(data[['x1','x2']])
est = sm.OLS(data[['y']], X2)
est2 = est.fit()
print(est2.summary())
```
