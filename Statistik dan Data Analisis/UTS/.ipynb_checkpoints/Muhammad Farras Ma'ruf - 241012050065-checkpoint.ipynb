```python
import pandas as pd
import numpy as np
from scipy import stats
```

## Datasets

Sumber dataset saya ambil dari kaggle.com, dimana provider menyediakan data mahasiswa dari Perguruan Tinggi XYZ. Banyak data sebesar 134 baris yang terdiri dari 4 kolom, diantaranya;

1. Index
2. Nama
3. UTS
4. UAS

Unuk kalkulasi dalam mengetahui **pemusatan datam**, **penyebaran data**, dan **analisa pendugaan** menggunakan rata-rata nilai dari UTS dan UAS. Maka dari itu kita perlu menambah kolom baru data asli untuk mengetahui nilai rata-rata UTS dan UAS yang ditujukan dengan nama kolom _**final_point**_.

```python
data_set = pd.read_csv("Data Nilai Mahasiswa.csv", sep=";")

banyak_data = len(data_set)

# Merubah tipe data kedalam float
data_set['UTS'] = data_set['UTS'].str.replace(",",".").astype(float)
data_set['UAS'] = data_set['UAS'].str.replace(",",".").astype(float)

# Menghitung nilai rata-rata uts dan uas
data_set['final_point'] = (data_set.UTS + data_set.UAS) / 2

data_set
```

# Jawaban 1

## Jawaban 1.a

Berbicara tentang pemusatan data, yang saya dapati pada matakuliah pak Tukiyat dalam materi statistik deskriptif ada 3, pemusatan data dapat dilihat dari nilai tengah (median), banyaknya data (modus) dan nilai tengah (mean). Dengan megetahui 3 nilai tersebut kita dapat membuat kesimpulan pemusatan dari data tersebut.

berikut adalah penyelesaian menggunakan fungsi numpy, namun secara konsep saya menguasaia untuk mendapatkan nilai Mean, Median dan Modus.

```python
median = np.median(data_set.final_point)
mean = np.mean(data_set.final_point)

# Modus
modus = stats.mode(data_set.final_point).mode
print(f"Nilai tengaah {median}")
print(f"Nilai rata {mean}")
print(f"Nilai intentsias paling banyak {modus}")
```

* Kedua angka ini (78.5 dan 80.0) sangat berdekatan dan relatif tinggi, menunjukkan bahwa sebagian besar mahasiswa memiliki performa yang baik.

### Adanya "Tarikan" Nilai Rendah:

Nilai rata-rata (Mean) di angka 76.65 secara signifikan lebih rendah daripada median (78.5) dan modus (80.0).

## Jawaban 1.b

Berbicara tentang pemusatan data, yang saya dapati pada matakuliah pak Tukiyat dalam materi statistik deskriptif, penyebaran dapat kita ketahui dengan mengetahui nilai dari `variance` dan `std deviasi`

berikut adalah penyelesaian menggunakan fungsi numpy, namun secara konsep saya menguasai untuk mendapatkan nilai varianec dan std variance.

Untuk mendapatkan nilai variance kita memerlukan kuadrat dari nilai dan nilai rata-rata lalu dibagi dengan `N` (banyaknya data)

```python
# Menyiapkan kuadrat dari selisih nilai dengan rata-rata nilai

## Membuat data set baru yg menampung parameter yg dibutuhkan untuk menghitung nilai variance
data_set_penyebaran = pd.DataFrame({'nilai_akhir': data_set.final_point})

## Membuat kolom selisih nilai dengan rata-rata nilia
data_set_penyebaran['selisih_nilai_d_rata_rata'] = data_set_penyebaran.nilai_akhir - mean

## Kuadrat nilai
data_set_penyebaran['pow_selisih'] = pow(data_set_penyebaran.selisih_nilai_d_rata_rata,2)
data_set_penyebaran.head()
```

```python
# Menghitung Variance (tanpa akar)
variance = sum(data_set_penyebaran.pow_selisih) / (banyak_data-1) 

# Standard deviasi adalah akar dari variance
standard_deviasi = np.sqrt(variance)

print(f"variance {variance}")
print(f"standard deviasi {standard_deviasi}")
```

### Analisis Penyebaran data dan data pemusatan diatas

Jika saya melihat dari daa diatas (mean, median dan modus) semua hasilnya berikut;abs

* Mean (Rata-rata): 76.65
* Median (Nilai Tengah): 78.5
* Modus (Nilai Terbanyak): 80.0
* Standard Deviasi (Sebaran): 9.17

Menurut hasil tersebut dan analisa saya performa Kelas Cukup Baik dan Homogen:
* Kelas ini memiliki performa yang baik (Median/Modus di ~78-80) dan nilainya cukup seragam (homogen).
* Standard deviasi 9.17 mengonfirmasi bahwa tidak ada variasi nilai yang ekstrem di antara mayoritas mahasiswa.
* Estimasi Sebaran Mayoritas: Dalam distribusi normal, sekitar 68% data berada dalam $\pm 1$ standard deviasi dari mean.
* Rentang: $76.65 \pm 9.17$Hasil Rentang: 67.48 s.d. 85.82Ini berarti, mayoritas (sekitar 68% atau $\approx$ 91 dari 134 mahasiswa) kemungkinan besar memiliki nilai gabungan UTS/UAS di antara 67.5 hingga 85.8.


Sehingga hasil penyebaran data justru memperkuat analisis skewed, dimana ini memperkuat analisis sebelumnya bahwa data condong ke kiri (skewed left). Mayoritas mahasiswa berkumpul di rentang nilai atas (67.5 - 85.8), sementara segelintir mahasiswa dengan nilai rendah (jauh di bawah 67.48) adalah "pencilan" yang menarik nilai rata-rata (Mean) ke bawah.

## Jawaban 1.c

```python
# Data yang Diketahui
x_bar = mean # nilai ratarata
s = standard_deviasi
n = banyak_data
cl = 0.90 ## Arahan pak tukiya 90% keyakinan

# Menghitung Standard Error (SE)
se = s / np.sqrt(n)

# Menghitung interval secara langsung
# 'confidence' = Tingkat Kepercayaan
# 'loc' = Titik pusat interval (Mean)
# 'scale' = Standar deviasi dari distribusi sampling (Standard Error)
ci_90 = stats.norm.interval(confidence=cl, 
                             loc=x_bar, 
                             scale=se)

print(f"--- Hasil Perhitungan Cepat (scipy) ---")
print(f"Standard Error: {se:.4f}")
print(f"Interval Kepercayaan 90%: {ci_90}")
print(f"Artinya, intervalnya adalah ({ci_90[0]:.3f} , {ci_90[1]:.3f})")
```

Analisa saya terkait pendugaan para meter, rata-rata dilakukan untuk mengestimasi nilai rata-rata populasi $\mu$ yang sebenarnya, melampaui 134 data sampel. Dengan menggunakan tingkat kepercayaan 90%, dan berdasarkan rata-rata sampel ($\bar{x}$) sebesar 76.65 serta standar deviasi (s) 9.17, didapatkan interval kepercayaan (75.35 , 77.95).

Interpretasi utamanya adalah: kita 90% yakin bahwa rata-rata nilai gabungan UTS/UAS yang sebenarnya untuk seluruh populasi mahasiswa berada di dalam rentang tersebut. Estimasi ini memiliki Margin of Error ($E$) sekitar $\pm 1.30$ poin, yang menunjukkan bahwa nilai rata-rata sampel (76.65) adalah penduga yang cukup presisi untuk rata-rata populasinya.

# Jawaban 2

```python
import scipy.stats as stats
import numpy as np

# --- Mendefinisikan Parameter Distribusi ---
# Ini adalah parameter populasi
mu = 250     # Rata-rata () waktu respons
sigma = 50   # Standar deviasi () waktu respons

print(f"Distribusi Normal Waktu Respons Server")
print(f"Rata-rata (): {mu} milidetik")
print(f"Standar Deviasi (): {sigma} milidetik\n")

# ==========================================================
## (a) Probabilitas waktu respons < 200 milidetik
# ==========================================================
# Kita mencari P(X < 200)

x_a = 200

# 1. Menghitung Z-score (Standardisasi)
z_a = (x_a - mu) / sigma
print(f"--- Soal (a): P(X < {x_a}) ---")
print(f"Z-score untuk {x_a} ms: {z_a:.2f}")

# 2. Menghitung probabilitas kumulatif
# Kita menggunakan cdf (Cumulative Distribution Function)
prob_a = stats.norm.cdf(x_a, loc=mu, scale=sigma)
# Alternatif: bisa juga pakai Z-score: prob_a = stats.norm.cdf(z_a)

print(f"Probabilitas P(X < {x_a}): {prob_a:.4f}")
print(f"Jadi, probabilitasnya adalah {prob_a*100:.2f}%\n")


# ==========================================================
## (b) Probabilitas waktu respons antara 150 ms dan 260 ms
# ==========================================================
# Kita mencari P(150 < X < 260)

x_b1 = 150
x_b2 = 260

# 1. Menghitung Z-score untuk kedua nilai
z_b1 = (x_b1 - mu) / sigma
z_b2 = (x_b2 - mu) / sigma
print(f"--- Soal (b): P({x_b1} < X < {x_b2}) ---")
print(f"Z-score untuk {x_b1} ms: {z_b1:.2f}")
print(f"Z-score untuk {x_b2} ms: {z_b2:.2f}")

# 2. Menghitung probabilitas
# P(150 < X < 260) = P(X < 260) - P(X < 150)
prob_b_atas = stats.norm.cdf(x_b2, loc=mu, scale=sigma)
prob_b_bawah = stats.norm.cdf(x_b1, loc=mu, scale=sigma)
prob_b = prob_b_atas - prob_b_bawah

print(f"\nP(X < {x_b2}) = {prob_b_atas:.4f}")
print(f"P(X < {x_b1}) = {prob_b_bawah:.4f}")
print(f"Probabilitas P({x_b1} < X < {x_b2}): {prob_b_atas:.4f} - {prob_b_bawah:.4f} = {prob_b:.4f}")
print(f"Jadi, probabilitasnya adalah {prob_b*100:.2f}%")
```

Menurut analisa saya berdasarkan distribusi normal waktu respons server ($\mu=250$ ms, $\sigma=50$ ms), ditemukan dua kesimpulan probabilitas. Pertama, probabilitas bahwa waktu respons server kurang dari 200 milidetik adalah 15.87%. Nilai 200 ms ini berada tepat -1.0 standar deviasi (Z-score = -1.0) di bawah rata-rata, yang mengindikasikan kejadian yang relatif jarang namun masih dalam ekspektasi wajar. Kedua, probabilitas waktu respons berada di antara 150 milidetik dan 260 milidetik adalah 55.65%. Rentang ini mencakup area dari -2.0 standar deviasi (untuk 150 ms) hingga +0.2 standar deviasi (untuk 260 ms) dari rata-rata. Perhitungan $P(150 < X < 260)$ ini menunjukkan bahwa lebih dari separuh permintaan pengguna akan ditangani dalam rentang waktu tersebut, menjadikannya skenario yang paling mungkin terjadi.
