# Distribusi Binomial

```python
import numpy as np
import pandas as pd
import math
from scipy import stats
```

![distribusi binom](./assets/distribusi_probabilitas_binomial.png)

Contoh : 

Kepada mahasiswa diberikan kesempatan untuk tidak masuk kuliah sebanyak 4 kali dari 10 kali pertemuan. Jika dalam 1 kelas ada 5 mahasiswa, maka peluang kelima mahasiswa tersebut masuk semua?

Diketehui :

$P = 4/10 = .4$ _probabilitas tidak masuk (siswa tidak masuk 4 kali) dibagi dengan jumlah masuk_ 
$P = 6/10 = .6$ _probabilitas masuk dibagi dengan jumlah masuk_ 

Pertanyaan :abs

Berapa peluang jika seluruh mahasiswa masuk, $ X = 10$

```python
jumlah_percobaan = 5
jumlah_sukses = 1 # kemungkinan yang tidak masuk 0
prob_sukses = .4
prob_gagal = 1 - prob_sukses
```

```python
n_fac = math.factorial(jumlah_percobaan)
x_fac = math.factorial(jumlah_sukses)
n_min_x_fac = math.factorial((jumlah_percobaan - jumlah_sukses))
pow_prob_sukes = math.pow(prob_sukses,jumlah_sukses)
pow_prob_failed = math.pow(prob_gagal, jumlah_percobaan - jumlah_sukses)

print(n_fac,x_fac,n_min_x_fac, pow_prob_sukes, pow_prob_failed)
```

```python
P_x = (n_fac / (x_fac * n_min_x_fac )) * pow_prob_sukes * pow_prob_failed
P_x * 100
```

```python
distribusi_binomial = pd.DataFrame({"sukses":range(0,5)})
distribusi_binomial['probabilitas'] = [stats.binom.pmf(a,5,0.4) for a in distribusi_binomial.sukses]
distribusi_binomial
```

```python
sum ( (distribusi_binomial.probabilitas))
```

## Another case

Sebuah perusahaan produk kosmetik ingin meluncurkan produk baru. Berdasarkan survei, perusahaan memprediksi bahwa 60% dari konsumen akan menyukai produk tersebut. Jika perusahaan melakukan uji coba dengan memberikan produk kepada 10 konsumen, hitunglah peluang bahwa
Ada  6 konsumen akan menyukai produk tersebut.

```python
distribusi_binomial2 = pd.DataFrame({"sukses":range(0,10)})
distribusi_binomial2['probabilitas'] = [stats.binom.pmf(a,10,0.6) for a in distribusi_binomial2.sukses]
distribusi_binomial2
```

```python
sum ( (distribusi_binomial2.probabilitas))
```

# Distribusi Poisson

![distribusi poisson](./assets/distribusi_poisson.png)

case : 

Apabila diketahui probabilitas seseorang akan meninggal dunia karena terkena penyakit anjing gila adalah 0,01. Sementara itu rata-rata orang yang meninggal akibat menderita penyakit anjing gila adalah 2 orang. Hitunglah peluang untuk :
* 3 orang akan meninggal
* Tidak lebih dari 1 orang yang meninggal
* Lebih dari 3 orang meninggal

```python
df_possion = pd.DataFrame({"data":range(0,4)})
df_possion['cummulative'] = [ stats.poisson.pmf(a,3) for a in df_possion.data] 
df_possion
```

```python
sum (df_possion.cummulative)
```
