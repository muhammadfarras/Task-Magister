## Jawaban Soal 1 : A

$H_0$ (Hipotesis Nol): $Median = 3$ (Tidak ada perbedaan yang signifikan antara median tingkat kepuasan mahasiswa dengan nilai standar 3).

$H_1$ (Hipotesis Alternatif): $Median \neq 3$ (Terdapat perbedaan yang signifikan antara median tingkat kepuasan mahasiswa dengan nilai standar 3).

## Jawaban Soal 1 : B (Uji statistik)

```python
stat_uji, p_value = stats.wilcoxon(data_soal1.selisih, alternative='two-sided')
median_aktual = np.median(data_soal1.skor)
```

## Jawaban Soal 1 : C dan D (Hasil uji dan interpretasi)

```python
print(f"Median Sampel: {median_aktual}")
print(f"Nilai Standar: {3}")
print("-" * 30)
print(f"Nilai Statistik Uji (W): {stat_uji}")
print(f"Nilai P-Value: {p_value:.5f}")
```

Menurut perhitungan yang saya lakukan diatas, maka;

1. Nilai Statistik Uji (W):Nilai ini merupakan jumlah terkecil dari ranking bertanda (signed rank). Dalam contoh simulasi di atas, nilai W kemungkinan kecil karena data condong positif (jauh dari netral).Contoh Hasil: **12.0**
2. Nilai P-Value:Probabilitas memperoleh hasil ekstrem jika $H_0$ benar.Hasil: **0.000153** (Sangat kecil)
3. Keputusan Uji:Kita membandingkan P-value dengan tingkat signifikansi ($\alpha$), biasanya 0.05 (5%).Aturan Keputusan:
* Jika P-value < 0.05 $\rightarrow$ Tolak $H_0$.
* Jika P-value > 0.05 $\rightarrow$ Gagal Tolak $H_0$.

Dalam simulasi di atas: Karena $000153 < 0.05$, maka Tolak $H_0$.

Dari hasil uji statistik Wilxocon One Sample diatas melihat apakah statndar kepuasan = 3 atau tidak, akan tetapi kita tidak mengetahui araah-nya. Karena nilai tersebut tidak sama dnegan 3, dengan membandingkan dengan nilai media sample dengan nilai **4** yang mana lebih tinggai dari nilai 3 maka dapat disimpulkan _**kinerja layanan akademik perguruan tinggi tersebut dinilai baik dan memuaskan oleh mahasiswa secara signifikan**_.

## Soal 2



### Menyiapkan Data

```python
# Memuat data
df_gabungan = pd.read_csv('./data/data_2.csv')
data_laki = np.array([3, 4, 2, 4, 3, 5, 2, 3, 4, 3, 2, 4])
data_perempuan = np.array([5, 4, 5, 4, 4, 5, 3, 5, 4, 5, 4, 3])
df_gabungan = df_gabungan.drop(columns=['Unnamed: 0'])
df_gabungan.head()
```

```python
df_gabungan.describe().head()
```

DataFrame yang digunakan pada uji **Mann-Whitney U (Dua Kelompok Independen)** terdiri dari 4 baris data yang terdiri dari informasi Nama Mahasiswa, jenis kelamin dan nilai kepuasan dengan skor tingkat kepuasan dengan rentang nilai 1-5 dimana 1 tidak puas hingga 5 sangat puas.

## Jawaban Soal 2 : A

Variable dependen disini adalah **Skor kepusan** sedangkan variable independen adalah jenis **kelamin mahasiswa**. Dengan detail jawaban saya sebagai berikut;abs

**Variabel Independen**
* Adalah: Jenis Kelamin (Gender)
* Kategori: Laki-laki dan Perempuan.

**Variabel Dependen**
* Adalah: Tingkat Kepuasan Mahasiswa
* Data: Skor Skala Likert (1 - 5).

## Jawaban Soal 2 : B (Uji Mann-Whitney)

```python
stat_u, p_value = stats.mannwhitneyu(data_laki, data_perempuan, alternative='two-sided')
```

```python
# Menghitung Z-score secara manual dari nilai U
n1 = len(data_laki)
n2 = len(data_perempuan)

# Mean ekspektasi U
mu_u = (n1 * n2) / 2

# Standar deviasi U
sigma_u = np.sqrt((n1 * n2 * (n1 + n2 + 1)) / 12)

# Menghitung Z
z_score = (stat_u - mu_u) / sigma_u
```

## Jawaban Soal 2 : C (Menentukan nilai Z dan p-value)

```python
print("\n--- Hasil Uji Mann-Whitney ---")
print(f"Statistik U: {stat_u}")
print(f"P-Value    : {p_value:.5f}")

print("\n--- Perhitungan Detail ---")
print(f"Nilai Z      : {z_score:.4f}")
print(f"Nilai P-Value: {p_value:.5f}")
```

## Jawaban Soal 2 : D (Interpretasi)

Dari hasi P-Value uji statistik diatas kita mendapatkan nilai **0.01580** dan bandingkan dengan Alpha (0.05).

Dengan hipotesa yang telah saya definisikan diatas;
* $H_0$: Tidak terdapat perbedaan signifikan pada tingkat kepuasan antara mahasiswa laki-laki dan
perempuan.
* $H_1$: Terdapat perbedaan signifikan pada tingkat kepuasan antara mahasiswa laki-laki dan perempuan.

Karena P-value **0.01580** < 0.05 :maka: Tolak $H_0$.Interpretasi: Terdapat perbedaan tingkat kepuasan yang signifikan antara mahasiswa laki-laki dan perempuan. Jenis gender tidak memliki pengaruh tingkat kepuasan karena bisa jadi apa yang diberikan oleh kampus dari segi fasilitas dan pendidikan pada mahasiswa laki=laki atau perempuan sama rata.

### Menyiapkan data

```python
df_pembelajaran = pd.read_csv('./data./daa_3.csv')
df_pembelajaran = df_pembelajaran.drop(columns=['Unnamed: 0'])
df_pembelajaran.head()
```

## Jawaban Soal 3 : A (Alasan Penggunaan Uji KruskalWallis)
Uji Kruskal-Wallis digunakan karena alasan berikut:

1. Tipe Data Ordinal: Data kepuasan diukur menggunakan Skala Likert (1-5), yang merupakan data ordinal (berjenjang), bukan interval/rasio murni. Oleh karena itu, uji parametrik seperti ANOVA (Analysis of Variance) kurang tepat.

2. Lebih dari Dua Kelompok: Kita membandingkan 3 kelompok independen (Tatap Muka, Daring, Hybrid). Jika hanya 2 kelompok, kita menggunakan Mann-Whitney; jika lebih dari 2, kita gunakan Kruskal-Wallis.

3. Distribusi Data: Uji ini adalah uji non-parametrik yang tidak mewajibkan data berdistribusi normal (bebas distribusi).

## Jawaban Soal 3 : B (Rumusan Hipotesa)

* $H_0$ (Hipotesis Nol): Median tingkat kepuasan mahasiswa sama pada ketiga metode pembelajaran (Tidak ada perbedaan signifikan antar metode).
* $H_1$ (Hipotesis Alternatif): Minimal ada satu metode pembelajaran yang memiliki median tingkat kepuasan yang berbeda (Terdapat perbedaan signifikan antar metode).

## Jawaban Soal 3 : C & D (Pengujian dan Kesimpulan)

```python
# Menyiapkan data per kelompok untuk fungsi kruskal
grup_tm = df_pembelajaran[df_pembelajaran['Metode'] == 'Tatap Muka']['Skor Kepuasan']
grup_daring = df_pembelajaran[df_pembelajaran['Metode'] == 'Daring']['Skor Kepuasan']
grup_hybrid = df_pembelajaran[df_pembelajaran['Metode'] == 'Hybrid']['Skor Kepuasan']

# Melakukan Uji Kruskal-Wallis
stat_h, p_value = stats.kruskal(grup_tm, grup_daring, grup_hybrid)

print(f"Nilai H-Statistic: {stat_h:.4f}")
print(f"Nilai P-Value    : {p_value:.5f}")

# Logika Pengambilan Keputusan Otomatis
alpha = 0.05
print("\n--- KESIMPULAN ---")
if p_value < alpha:
    print("Keputusan: Tolak H0")
    print("Interpretasi: Terdapat perbedaan tingkat kepuasan yang SIGNIFIKAN antara metode Tatap Muka, Daring, dan Hybrid.")
else:
    print("Keputusan: Gagal Tolak H0")
    print("Interpretasi: TIDAK terdapat perbedaan yang signifikan (tingkat kepuasan dianggap sama).")
```

Nilai P-Value: **0.00159**, jauh di bawah 0.05. Sehingga dapat saya simpulkan: Karena $P-value < 0.05$, maka Menolak $H_0$.Implikasi: Secara statistik terbukti bahwa metode pembelajaran mempengaruhi kepuasan mahasiswa.
